```python
import numpy as np
import matplotlib.pyplot as plt

import sympy as sym
import sympy.plotting.plot as symplot
```

```python
# symbolic variable
x = sym.symbols('x')

fx = 2*x**2
df = sym.diff(fx, x)

# compute derivative of the function
print(df)
```

```python
symplot(fx,(x,-4,4),title='function')
plt.show()
symplot(df,(x,-4,4),title='derivative')
plt.show()
```

```python
# with ReLU and sigmoid

relu = sym.Max(0,x)
sigmoid = 1 / (1 + sym.exp(-x))

p = symplot(relu,(x,-5,2),label='ReLU', show=False,line_color='blue')
p.extend(symplot(sigmoid,(x,-5,5),label='sigmoid',show=False,line_color='red'))
p.legend = True
p.show()

p = symplot(sym.diff(relu),(x,-5,5),label='df(ReLU)', show=False,line_color='blue')
p.extend(symplot(sym.diff(sigmoid),(x,-5,5),label='df(sigmoid)',show=False,line_color='red'))
p.legend = True
p.show()
```

# Chain rule

```python
from IPython.display import display
```

```python
a = sym.symbols('a')
fx = 2*a**2
gx = 4*a**3 - 3*a**4

df = sym.diff(fx, a)
dg = sym.diff(gx, a)

print('Functions')
display(fx)
display(gx)
print('\nTheir derivatives')
display(df)
display(dg)
print('\nProduct rule')
display(df * gx + fx * dg)

print('\nWith sympy')
display(sym.diff(fx * gx, a))
```
