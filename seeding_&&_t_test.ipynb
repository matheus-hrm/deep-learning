# Seeding

A seed purpose is to be able to generate the same set of random numbers multiple times, so this code can be reproducible.

```python
# sets the seed
seed = np.random.RandomState(3)

# generates a array of randomly picked 5 numbers
print( seed.randn(5) )

# it will generate these same 5 numbers regardless
# [ 1.78862847  0.43650985  0.09649747 -1.8634927  -0.2773882 ]
```

# T test

Comparação de Modelos:

---


Permite verificar se diferenças entre métricas de desempenho (e.g., acurácia, F1-Score) de dois modelos são estatisticamente significativas, evitando conclusões baseadas em variações aleatórias.

O teste T assegura que decisões (e.g., escolha de modelos, features) sejam respaldadas por evidências estatísticas, reduzindo riscos de overfitting e viés em conclusões experimentais.

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# samples for datasets
n1 = 30
n2 = 40

# population mean in the datasets
mu1 = 1
mu2 = 2

data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

plt.plot(np.zeros(n1),data1,'ro',markerfacecolor='w',markersize=14)
plt.plot(np.ones(n2),data2,'bs',markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group1','Group2'])
plt.show()
```

```python
# t-test

t,p = stats.ttest_ind(data1, data2)
print(t)
print(p)
```

```python
fig = plt.figure(figsize=(10,4))
plt.plot(0+np.random.randn(n1)/15,data1,'ro',markerfacecolor='w',markersize=14)
plt.plot(1+np.random.randn(n2)/15,data2,'bs',markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group1','Group2'])
plt.show()
```
