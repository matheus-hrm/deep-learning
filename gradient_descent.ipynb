```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
def f(x):
  return 3*x**2 - 3*x + 4

def deriv(x):
  return 6*x - 3

x = np.linspace(-2,2,2001)

plt.plot(x,f(x),x,deriv(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','f\'(x)'])
plt.show()
```

```python
# random start
localmin = np.random.choice(x,1)
print(localmin)

# parameters
# when you have a slow learning rate you need to adjust to a higher number of epochs to enable
# the model to learn more slowly, e.g. rate = .001 epochs = 1000
learning_rate = .03
epochs = 100

for epoch in range(epochs):
  grad = deriv(localmin)
  localmin = localmin - learning_rate*grad

print(localmin)
```

```python
l = np.random.choice(x,1)
print(l)

rate = .01
epochs = 100

params = np.zeros((epochs,2))
for epoch in range(epochs):
  grad = deriv(l)
  l = l - rate*grad
  params[epoch,0] = l.item()
  params[epoch,1] = grad.item()
```

```python

fig, ax = plt.subplots(1,2,figsize=(10,5))

for i in range(2):
  ax[i].plot(params[:,i],"o-")
  ax[i].set_xlabel('iteration')
  ax[i].set_ylabel(f'final minimum {l[0]:.5f}')

ax[0].set_ylabel('f(x)')
ax[1].set_ylabel('f\'(x)')

plt.show()
```

```python
# example
pi = np.pi
def function(x):
  return np.cos( 2 * pi * x) + (x ** 2)

def derivative(x):
  return -2 * pi * np.sin(2 * pi * x) + 2 * x

x = np.linspace(-2,2,2001)

local = np.random.choice(x,1)
print(local)

rate = .01
epochs = 1000

for i in range(epochs):
  grad = derivative(local)
  local = local - rate*grad

print(local)
plt.plot(x,function(x),x,derivative(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','f\'(x)'])
plt.show()
```
